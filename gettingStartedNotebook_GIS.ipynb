There are different ways to connect with the server. Here we create a "Connection" instance:

```python
conn = connect(user='XXXXXXXX', host='lav-fileserver', port='5433', dbname='mobility')
```

In the notebook we can write and execute the same SQL queries as in pgAdmin, although with few adapatations:

```python
sql = ['SELECT *',
    'FROM swissboundaries.swissboundaries3d_1_3_tlm_2017_kantonsgebiet']

sql = "\n ".join(sql)
df = pd.read_sql(sql=sql, con=conn, index_col='gid')
```

Here `sql` is initally a list (comma separated), where each element is a string of the query. For human readability you can write a string for each line of the query. Then all elements of the list are merged together with the `join` command, i.e. sql becomes a single string with the entire query in it. The last command executes the query through the server connection `conn` and returns the output as a pandas DataFrame:

```python
df
```

This is the same table output as in pgAdmin: 51 rows, all attributes and the explicit **geom** column. By importing the data with the pandas command `pd.read_sql()` we are not including any geographical meaning to the table. If we do:

```python
df.loc[32,'geom']
```

we get a very long string that would represent the geometry. To convey the geographical meaning we need to import the ouptu of the query through geopandas commands:

```python
sql = ['SELECT *,',
      'ST_MakeValid(geom) as valid_geom',
      'FROM swissboundaries.swissboundaries3d_1_3_tlm_2017_kantonsgebiet']

sql = "\n ".join(sql)
df = gpd.GeoDataFrame.from_postgis(sql=sql, con=conn, index_col='gid', geom_col='valid_geom')
```

`ST_MakeValid` is a PostGIS function required to "fix" (if possible) a geometry that wasn't easily readable (without it you get an error). Note that we explicitly set `geom_col='valid_geom'`, since otherwise the default column **geom** would be chosen and the latter has the "broken" geometry.

```python
type(df)
```

```python
df.loc[32,'valid_geom']
```

```python
df['valid_geom']
```

All the last commands show that the **valid_geom** column is correctly interpreted as a geometry. The last command also shows the geometry in a different format: **Well-known text (WKT)**. Contrary to the what could be seen in the raw table (**Well-known binary (WKB)**), the geometries have now a human-redable format, with the numbers representing physical coordinates (specifically in CH1903+ coordinate reference system).

Nice explanation for geometry types and formats here: https://www.gaia-gis.it/gaia-sins/spatialite-cookbook/html/wkt-wkb.html

Read more about the Swiss coordinate systems here: https://en.wikipedia.org/wiki/Swiss_coordinate_system

## Plots

The fastest way is

```python
df.plot();
```

With some refinement:

```python
fig, ax = plt.subplots(figsize=(12,8))     # for the size and have
df.plot(ax=ax, column='einwohnerz')    
ax.set_aspect('equal')   # set aspect to equal. This is done automatically with *geopandas* plots, but not when working with pyplot.
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off');      # to make the "border frame", i.e. the axes, disappear
```

We are now plotting all the (51!) shapes of `df` with a color related to the column/attribute **einwohnerz** (population) for each shape. It looks nice but there is a problem: different parts belonging to the same canton (like Vaud, Fribourg or Bern) have different colors because of different information in the **einwohnerz** column. By checking the DataFrame for shapes belonging to canton Bern:

```python
df[df['name']=='Bern']
```

we see that **einwohnerz** attribute is provided only to the main part of canton Bern, not to all its other portions. The presence of `NaN` values in the **einwohnerz** column is also the reason for the _RuntimeWarning_ raised above.

The solution is to employ the query written in SQL that merged geometries. We will update the query to extract also information about population and area (**kantonsfla** attribute). As mentioned in the PostGIS tutotial, to extract additional columns from a query that groups rows, we need to assign an aggregate functions to such columns. In this specific case, we want to extract the only row of each group that is not `Nan`. The SQL aggregate function **max()** is good enough for this purpose.

```python
sql = ['SELECT kantonsnum,',
    'count(*) AS n_parts,',
    'ST_MakeValid(ST_Union(geom)) AS geom,',
    'max(einwohnerz) AS pop,',
    'max(kantonsfla) AS area',
    'FROM swissboundaries.swissboundaries3d_1_3_tlm_2017_kantonsgebiet',
    'GROUP BY kantonsnum']

sql = "\n ".join(sql)
df = gpd.GeoDataFrame.from_postgis(sql=sql, con=conn, index_col='kantonsnum', geom_col='geom')
```

Note that:
- we could apply PostGIS functions in series (this applies to any SQL function)
- since we don't extract the original **geom** column any more, we could rename the transformed geometry column exactly **geom**

```python
len(df)
```

```python
fig, ax = plt.subplots(figsize=(12,8))     # for the size and have
df.plot(ax=ax, column='pop')    
ax.set_aspect('equal')  
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
```

We can also plot the population density. Note that area is given in hectares ($100m$ x $100m$ = $10,000m^2$). This time we also add title and colorbar for readability.

```python
df['density'] = df['pop']/df['area']*100
```

```python
cmap = 'winter_r'  # colormap
alpha = 0.9        # transparency
vmin = 0
vmax = 900

fig, ax = plt.subplots(figsize=(15,8))     # for the size and have
df.plot(ax=ax, column='density', cmap=cmap, vmin=vmin, vmax=vmax, alpha=alpha)    
ax.set_aspect('equal')   
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
ax.set_title("Population density by canton", fontsize=18)

# To add colorbar: sm is a *fake* ScalarMappable
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cb = fig.colorbar(sm, ax=ax, shrink=0.9, alpha=alpha)
cb.ax.tick_params(labelsize=16)
cb.set_label('people/km^2', fontsize=16, labelpad=10)
```
